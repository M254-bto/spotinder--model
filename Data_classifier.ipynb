```python

def feature_extract():
  features = []
  for song in songs:
    uri = song['uri']
    song_features = sp.audio_features(uri)
    for feature in song_features:
      features.append(feature)
  return pd.DataFrame(features)

feature_extract()
```

```python

def prep():
  data = feature_extract()
  data.drop(['id', 'uri', 'track_href', 'analysis_url', 'time_signature', 'type'], axis = 1, inplace = True) 
  
  return data
prep()
```
